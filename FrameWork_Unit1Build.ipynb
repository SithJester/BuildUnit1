# Healthcare Industry
Cleaning DF's to isolate high probability of automation by occupation, showing impact on each state vs. total employees per occupation

```
## Isolate the Healthcare industry by job code

Medical = State_Autom[State_Autom['SOC'].str.contains('29-')]
```

```
Medical.head()
```

```
# Isolate the healthcare industry salary by job code

Med_Salary = Salary[Salary['OCC_CODE'].str.contains('29-')]
```

```
Med_Salary
```

```
## Sum all of the employees by state into a dataframe

TotalMedEmp = pd.DataFrame(np.round(Medical.sum(axis=1)),columns={'TOT_EMP'})
```

```
TotalMedEmp.head()
```

```
## Add the total employee count to the healthcare by state DF

Health = pd.concat((TotalMedEmp,Medical),axis=1
          )

print(Health.shape)
Health.head()
```

```
Health.loc['All_EMP'] = pd.Series(Health['TOT_EMP'].sum(), index = ['TOT_EMP'])
```

```
## Isolate the job titles that have a high probablility of automation


HProb_Med = Health[Health['Probability']>=.5]
```

```
HProb_Med
```

```
## Make a list of the state columns
#  to get totals by state per high probability

col_list = list(HProb_Med)
col_list.remove('SOC')
col_list.remove('TOT_EMP')
col_list.remove('Occupation')
col_list.remove('Probability')
col_list
```

```
HProb_Med.loc['State_EMP'] = HProb_Med[col_list].sum()
```

```
HProb_Med.loc['All_EMP'] = pd.Series(HProb_Med['TOT_EMP'].sum(), index = ['TOT_EMP'])
```

```
print(HProb_Med.shape)
HProb_Med
```
